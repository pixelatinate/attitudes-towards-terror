```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import csv
  
x = []
y = []
  
with open('GallupTerrorismDataset.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter = ',')
    for row in plots:
        x.append(row[0])
        y.append(int(row[2]))

plt.bar(x, y, width = 0.72, label = "Percent Positive",color='#76b800')
plt.xlabel('Years')
plt.ylabel('Percent Positive')
plt.title('Thinking now about U.S. military action in Afghanistan that began in October 2001, do you think the United States made a mistake in sending military forces to Afghanistan, or not?')
plt.ylim(0,100)
plt.gca().xaxis.set_major_locator(MaxNLocator(prune='lower'))
plt.legend()
plt.show()
```
